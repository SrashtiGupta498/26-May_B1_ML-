```python
import tkinter as tk
import pandas as pd
from tkinter import * 
from tkinter.ttk import *
import pandas as pd
from tkinter import messagebox

app = tk.Tk()
app.title('Registration Form')
app.geometry('400x280')
app.configure(bg='lightgray')
app.configure(bg='pink')

f_name = tk.Variable(app)
e_mail = tk.Variable(app)
u_name = tk.Variable(app)
pwd = tk.Variable(app)
age = tk.Variable(app)
mobile = tk.Variable(app)

tk.Label(app,text='Full Name',bg='black',fg='white',padx=1,pady=1,font='Algerian''bold').place(x=20,y=30)
tk.Label(app,text='Email ID',bg='black',fg='white',font='Algerian''bold',padx=2,pady=1).place(x=20,y=60)
tk.Label(app,text='User Name',bg='black',fg='white',padx=1,pady=1,font='Algerian''bold').place(x=20,y=90)
tk.Label(app,text='Password',bg='black',fg='white',padx=1,pady=1,font='Algerian''bold').place(x=20,y=120)
tk.Label(app,text='Age',bg='black',fg='white',padx=1,pady=1,font='Algerian''bold').place(x=20,y=150)
tk.Label(app,text='Mobile',bg='black',fg='white',padx=1,pady=1,font='Algerian''bold').place(x=20,y=180)

tk.Entry(app,width=30,fg='black',bg='white',textvariable=f_name).place(x=150,y=30)
tk.Entry(app,width=30,fg='black',bg='white',textvariable= e_mail).place(x=150,y=60)
tk.Entry(app,width=30,fg='black',bg='white',textvariable= u_name).place(x=150,y=90)
tk.Entry(app,width=30,fg='black',bg='white',textvariable=pwd,show='*').place(x=150,y=120)
tk.Entry(app,width=30,fg='black',bg='white',textvariable=age).place(x=150,y=150)
tk.Entry(app,width=30,fg='black',bg='white',textvariable=mobile).place(x=150,y=180)

def error():
    app1=tk.Toplevel(app)
    app1.geometry('150x94')
    messagebox.showerror("showerror", "Error")

def register():
    Mobile=mobile.get()
    Age=age.get()
    Password=pwd.get()
    if len(Mobile)>10:
        return error()
    elif len(Age)>2:
        return error()
    elif len(Password)>8:
        return error()
    else:
        tk.Label(app,text='User has been registered succesfully').place(x=15,y=230)
        
        f = open('data.csv','a')
        data = [f_name.get(),e_mail.get(),u_name.get(),pwd.get(),age.get(),mobile.get()]
        f.write(','.join(data)+'\n')

        f.close()
        f_name.set('')
        e_mail.set('')
        u_name.set('')
        pwd.set('')
        age.set('')
        mobile.set('')



        cols=('Full Name','Email ID','User Name','Password','Age','Mobile')
        df=pd.read_csv('Data.csv',names=cols)
        new_df = df.dropna()

        new_window = tk.Toplevel(app)
        new_window.title("Welcome to new Window")
        new_window.geometry("400x280")
        Label(new_window,text ="This is a new window")

        total_user_var=tk.Variable(app)
        Average_age_var=tk.Variable(app)
        mail_domain_var=tk.Variable(app)
        google_var=tk.Variable(app)
        yahoo_var=tk.Variable(app)
        gmail_var=tk.Variable(app)

        total_user_var.set(df['Full Name'].count())
        Average_age_var.set(df['Age'].mean())
        mail_domain_var.set(new_df['Email ID'].count())

        mail=new_df['Email ID']
        google_var.set(new_df['Email ID'][mail.str.endswith('@google.com')].count())
        yahoo_var.set(new_df['Email ID'][mail.str.endswith('@yahoo.com')].count())
        gmail_var.set(new_df['Email ID'][mail.str.endswith('@gmail.com')].count())


    
    
        tk.Label(new_window,text='SUMMARY',bg='red',foreground='white',font=('Algerian', 15, 'bold')).place(x=130,y=15)
        total_user=tk.Label(new_window,text='Total Users:     ',bg='#f5ffff',font=('Algerian', 10, 'bold')).place(x=20,y=60)
        Average_age=tk.Label(new_window,text='Average age:     ',bg='#f5ffff',font=('Algerian', 10, 'bold')).place(x=20,y=90)
        mail_domain=tk.Label(new_window,text='Commonly mail Domain: ',bg='#f5ffff',font=('Algerian', 10, 'bold')).place(x=20,y=130)
        google=tk.Label(new_window,text='Google.com: ',bg='#f5ffff',font=('Algerian', 10, 'bold')).place(x=20,y=160)
        yahoo=tk.Label(new_window,text='Yahoo.com: ',bg='#f5ffff',font=('Algerian', 10, 'bold')).place(x=20,y=190)
        gmail=tk.Label(new_window,text='Gmail.com: ',bg='#f5ffff',font=('Algerian', 10, 'bold')).place(x=20,y=220)



        total_user=tk.Label(new_window,text='Total Users:',font=('Algerian', 10, 'bold'),textvariable=total_user_var).place(x=140,y=60)
        Average_age=tk.Label(new_window,text='Average age: ',font=('Algerian', 10, 'bold'),textvariable=Average_age_var).place(x=140,y=90)
        mail_domain=tk.Label(new_window,text='Commonly mail Domain: ',font=('Algerian', 10, 'bold'),textvariable=mail_domain_var).place(x=200,y=130)
        google=tk.Label(new_window,text='Google.com: ',font=('Algerian', 10, 'bold'),textvariable=google_var).place(x=120,y=160)
        yahoo=tk.Label(new_window,text='Yahoo.com: ',font=('Algerian', 10, 'bold'),textvariable=yahoo_var).place(x=120,y=190)
        gmail=tk.Label(new_window,text='Gmail.com: ',font=('Algerian', 10, 'bold'),textvariable=gmail_var).place(x=120,y=220)

    
    
tk.Button(app, text='Submit',bg='green',foreground='white',font=('Algerian',10,'bold'),command=register).place(x=150,y=210)



app.mainloop()
```
